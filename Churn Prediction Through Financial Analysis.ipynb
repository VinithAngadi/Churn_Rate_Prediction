# Data Preprocessing

```python
from sklearn.impute import SimpleImputer
from numpy import nan

# Imputing missing values
imputer = SimpleImputer(missing_values=nan, strategy='mean')
dataset['credit_score'] = imputer.fit_transform(dataset['credit_score'].to_numpy().reshape(-1, 1))
dataset['age'] = imputer.fit_transform(dataset['age'].to_numpy().reshape(-1,1))
```

```python
dataset = dataset[dataset.credit_score >= 300]
dataset = dataset.dropna()
```

```python
# dropping highly correlated features
dataset_new = dataset.drop(columns = ['app_web_user', 'deposits'])
```

```python
user_identifier = dataset_new['user']
dataset_new = dataset_new.drop(columns = ['user'])
```

```python
# One hot encoding
dataset_new = pd.get_dummies(dataset_new)
```

```python
dataset_new.columns
```

```python
dataset_new = dataset_new.drop(columns = ['housing_na', 'zodiac_sign_na', 'payment_type_na'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(dataset_new.drop(columns = 'churn'), dataset_new['churn'],
                                                    test_size = 0.3,
                                                    random_state = 0)
```

```python
import random

pos_index = Y_train[Y_train.values == 1].index
neg_index = Y_train[Y_train.values == 0].index

if len(pos_index) > len(neg_index):
    higher = pos_index
    lower = neg_index
else:
    higher = neg_index
    lower = pos_index

random.seed(0)
higher = np.random.choice(higher, size=len(lower))
lower = np.asarray(lower)
new_indexes = np.concatenate((lower, higher))

X_train = X_train.loc[new_indexes,]
Y_train = Y_train[new_indexes]
```

```python
Y_train.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
# Feature scaling 

sc_X = StandardScaler()
X_train2 = pd.DataFrame(sc_X.fit_transform(X_train))
X_test2 = pd.DataFrame(sc_X.transform(X_test))
X_train2.columns = X_train.columns.values
X_test2.columns = X_test.columns.values
X_train2.index = X_train.index.values
X_test2.index = X_test.index.values
X_train = X_train2
X_test = X_test2
```
